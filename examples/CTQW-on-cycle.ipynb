### Continuous-time quantum walk (CTQW) on a cycle

```python
import hiperwalk as hpw
```

```python
N = 101
cycle = hpw.Cycle(N)

# ctqw is an instance of a CTQW
ctqw = hpw.ContinuousTime(cycle, gamma=0.35)

# initial state
psi0 = ctqw.ket(N // 2)

psi_final = ctqw.simulate(time=50, state=psi0)

prob = ctqw.probability_distribution(psi_final)

hpw.plot_probability_distribution(prob,figsize=(5,3))
```
