## Perfect state transfer by CTQW on a hypercube

```python
import hiperwalk as hpw
import numpy as np
```

### Defining the graph and a QW instance

The next command creates a hypercube with name 'cube', which is an instance of the Graph class. The dimension of the hypercube in 'n'.

```python
n = 5
cube = hpw.Hypercube(n)
```

The next command creates a QW on the hypercube, which is an instance of the ContinuousTime class. We will take $\gamma=1$ and set $t=\pi/20$, which implies that $U=\exp(i\pi A/20)$.

```python
qw = hpw.ContinuousTime(cube, gamma=1, time=(np.pi/2)*1/10)
```

### Setting the initial state

```python
psi0 = qw.ket(0)
```

### Probability distribution at $t=\frac{\pi}{2}$ 

We expect PST from 0 to $2^n-1$ at $t=\pi/2$. We will plot the probability distribution at $\pi/2$ and we will check that the probability of finding the walker on vertex $2^n-1$ is 1.

```python
psi_final = qw.simulate(range=(10,11), state=psi0)
prob = qw.probability_distribution(psi_final)
hpw.plot_probability_distribution(prob, plot='histogram', figsize=(6,3))
```

```python
prob[0].tolist()[-1] # last entry of prob[0] 
```

### Animation of the quantum walk

```python
list_of_states = qw.simulate(range = (0,21), state = psi0)
list_of_probs = qw.probability_distribution(list_of_states)
hpw.plot_probability_distribution(list_of_probs, graph=cube, animate=True, figsize=(7,3))
```
