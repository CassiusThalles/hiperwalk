### Coined quantum walk on a hypercube

```python
import hiperwalk as hpw
import numpy as np
```

```python
# dimension of the hypercube
dim = 3

# cube is an instance of a hypercube
cube = hpw.Hypercube(dim)

# cqw is an instance of a coined QW on a hypercube
cqw = hpw.Coined(cube)
```

There are two ways to define the initial state: 

$$|\psi_0\rangle = \frac{1}{\sqrt{dim}}\sum_{a\in A}|a\rangle,$$

where $A$ is the set of arcs whose tails are vertex 0.

```python
# using the labels of the arcs: a is the label of an arc
# the range of a is (0,dim*2**dim-1)
ini_state = cqw.state([(1/np.sqrt(dim), a) for a in range(dim)])
print(ini_state)
```

```python
# using arcs (head,tail), where head and tail are vertices.
# the labels of the vertices are 0, 1, 2, 4, 8, 16, ..., 2**dim
ini_state = cqw.state([(1/np.sqrt(dim), (0,2**i)) for i in range(dim)])
print(ini_state)
```

```python
# simulate calculates the state at time steps
final_state = cqw.simulate(time = dim, state = ini_state)

probs = cqw.probability_distribution(final_state)

hpw.plot_probability_distribution(probs, graph=cube, figsize=(5,3))
```

```python
# alternative way of ploting the probability distribution
hpw.plot_probability_distribution(probs, plot='histogram', figsize=(5,3))
```
